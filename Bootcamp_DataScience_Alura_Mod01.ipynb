## Desafio 05: Buscar os casos de dengue no Brasil (época de maior número de casos e regiões mais atingidas) e se os picos de alguns estados em fevereiro e verão de modo geral, pode ser reflexos dos casos de dengue.

Utilizei uma base de dados dos casos no Brasil de 2008 a 2014

```
#Carregando a base de dados dos casos de dengue no Brasil
fonte_dengue = 'https://raw.githubusercontent.com/Raoni-Silva/BootcampAlura/main/Dengue_Brazil.csv'
dados_dengue = pd.read_csv(fonte_dengue, encoding='ISO-8859-1')
dados_dengue_2008_2014 = dados_dengue[18:]
dados_dengue_2008_2014 = dados_dengue_2008_2014.set_index('Year')
dados_dengue_2008_2014
```

```
#Criando a lista com os anos de interesse
ano = colunas_interessadas.columns.str.slice(start=0, stop=4).unique()[:-1]
ano
```

```
#Criando séries com as somas dos gastos por ano por estado
ano_2008 = colunas_interessadas.loc[: ,'2008/Fev':'2008/Dez'].sum(axis=1)
ano_2009 = colunas_interessadas.loc[: ,'2009/Jan':'2009/Dez'].sum(axis=1)
ano_2010 = colunas_interessadas.loc[: ,'2010/Jan':'2010/Dez'].sum(axis=1)
ano_2011 = colunas_interessadas.loc[: ,'2011/Jan':'2011/Dez'].sum(axis=1)
ano_2012 = colunas_interessadas.loc[: ,'2012/Jan':'2012/Dez'].sum(axis=1)
ano_2013 = colunas_interessadas.loc[: ,'2013/Jan':'2013/Dez'].sum(axis=1)
ano_2014 = colunas_interessadas.loc[: ,'2014/Jan':'2014/Dez'].sum(axis=1)
```

```
#Criação do DataFrame com os dados por ano
dados_2008_2014 = pd.DataFrame([ano_2008, ano_2009, ano_2010, ano_2011, ano_2012, ano_2013, ano_2014]).T
#Renomeando as colunas com os anos
dados_2008_2014.columns = ano[0:7]

x = dados_2008_2014.T.index
y1 = dados_2008_2014.mean()/1000000
y2 = dados_dengue_2008_2014['Cases']

#Criar a área de plotagem e os dois eixos y
fig, ax1 = plt.subplots(figsize=(10,6))
ax2 = ax1.twinx()

curva1, = ax1.plot(x, y1, label='Gasto por ano', color='r', marker=".", markersize=15)
curva2, = ax2.plot(x, y2, label='Casos de dengue', color='b',  marker=".", markersize=15)

#Criar uma lista para acessar os parametros das curvas
curvas = [curva1, curva2]

#Criar a legenda do gráfico
ax1.legend(curvas, [curve.get_label() for curve in curvas])

#Criar os labels de cada eixo
ax1.set_xlabel('Ano', {'size':15})
ax1.set_ylabel('Valores em milhões de reais', {'size':15}, color=curva1.get_color())
ax2.set_ylabel('Casos de dengue', {'size':15}, color=curva2.get_color())

#Alterar a cor dos eixos y
ax1.tick_params(axis='y', colors=curva1.get_color())
ax2.tick_params(axis='y', colors=curva2.get_color())

#Setar os valores dos eixos y
ax1.yaxis.set_major_formatter(ticker.StrMethodFormatter('R$ {x:,.0f}'))
ax2.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

plt.title('Comparativo anual de gastos\ncom saúde X Casos de dengue\n(Brasil)', {'size':15})
plt.show()
```
