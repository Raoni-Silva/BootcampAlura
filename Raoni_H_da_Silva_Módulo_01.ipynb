## 03 - Preparação dos bancos de dados

Criação da coluna de regiões do Brasil.

```
#Dicionário de regiões
regioes = {'SP': 'Sudeste', 'RJ': 'Sudeste', 'ES':'Sudeste', 'MG': 'Sudeste', 
           'PR':'Sul', 'RS':'Sul', 'SC': 'Sul',
           'AP': 'Norte', 'AC': 'Norte', 'AM': 'Norte', 'PA': 'Norte', 'RR':'Norte', 'RO': 'Norte', 'TO':'Norte',
           'AL':'Nordeste', 'CE':'Nordeste', 'BA':'Nordeste', 'MA':'Nordeste', 'PB': 'Nordeste', 'PE':'Nordeste', 'PI':'Nordeste', 'RN':'Nordeste', 'SE':'Nordeste',
           'DF':'Centro Oeste', 'MS':'Centro Oeste', 'MT':'Centro Oeste', 'GO':'Centro Oeste'}

#Criação da coluna de regiões
UF = covid_full_raw['state']
coluna_regiao = UF.map(regioes)

covid_full_raw['region'] = coluna_regiao
covid_full_raw.head()
```

Será utilizada apenas a população estimada de 2020 e portanto retirarei a coluna "estimated_population_2019" do dataframe.

```
covid_full = covid_full_raw.drop('estimated_population_2019', axis=1)
covid_full.head()
```

Baseado na coluna "last_available_confirmed_per_100k_inhabitants" criarei a coluna "deaths_per_100k_inhabitants".

```
covid_full['deaths_per_100k_inhabitants'] = (covid_full['last_available_deaths']/(covid_full['estimated_population']/100000))
covid_full.head()
```

Vou reordenar as colunas do dataframe de forma mais organizada.

```
colunas_ordenadas = ['date', 'last_available_date', 'region', 'state', 'city', 'place_type', 'order_for_place', 'epidemiological_week', 'new_confirmed', 'last_available_confirmed',  
                     'new_deaths', 'last_available_deaths', 'is_last', 'is_repeated', 'estimated_population', 'city_ibge_code', 'last_available_confirmed_per_100k_inhabitants', 
                     'deaths_per_100k_inhabitants', 'last_available_death_rate']
```

### Hipótese 05: os dados por município estão defasados com relação aos dados estaduais. 

```
casos_estado = covid_estado.query('is_last == True')[['state', 'last_available_confirmed']]
casos_estado = casos_estado.set_index('state')
casos_estado = casos_estado.rename(columns={'last_available_confirmed':'state_confirmed_cases'})
```

```
casos_municipio = covid_cidade.query('is_last == True').loc[:,['state','last_available_confirmed']].groupby('state').sum()
casos_municipio = casos_municipio.rename(columns={"last_available_confirmed": "city_confirmed_cases"})
```

```
casos_comparativo = pd.concat([casos_estado, casos_municipio], axis=1)
casos_comparativo['difference'] = casos_comparativo['state_confirmed_cases'] - casos_comparativo['city_confirmed_cases']
casos_comparativo['%_difference'] = round((1-(casos_comparativo['city_confirmed_cases'] / casos_comparativo['state_confirmed_cases']))*100, 2)
casos_comparativo
```

Conclusão da Hipótese 05: Existem alguns estados com dados municipais defasados com relação aos dados estaduais. Não saberia dizer qual seria o possível motivo disso.

Para evitar usar dados municipais com erros me concentrarei nas análises municipais apenas em municípios em que a porcentagem de dados faltantes seja menor que 1.0%

```
#Lista dos estados em que os dados municipais serão considerados em análises futuras.
estados_considerados = casos_comparativo[casos_comparativo['%_difference'] <= 1].index
estados_considerados
```

```
#Dataframe do comparativo da contagem de casos estaduais e municipais em que a diferença é inferior a 1%
casos_comparativo[casos_comparativo.index.isin(estados_considerados)]
```

```
#Criação do dataframe dos registros municipais 
covid_cidade_selecionados = covid_cidade[covid_cidade['state'].isin(estados_considerados)]
covid_cidade_selecionados.head()
```

```
covid_estado_selecionados = covid_estado.copy()
```

## 04 - Primeiras análises dos bancos de dados

```
#Conferindo o número de casos estaduais e municipais
casos_por_milhao_estado = round(covid_estado_selecionados.query('is_last == True')['last_available_confirmed'].sum()/1000000,2)
casos_por_milhao_cidade = round(covid_cidade.query('is_last == True')['last_available_confirmed'].sum()/1000000,2)
casos_por_milhao_cidade_selecionado = round(covid_cidade_selecionados.query('is_last == True')['last_available_confirmed'].sum()/1000000,2)
print('O número de registros estaduais de covid é de ', casos_por_milhao_estado, ' milhões de casos')
print('O número de registros municipais de covid é de ', casos_por_milhao_cidade, ' milhões de casos')
print('O número de registros municipais (de municípios de estados selecionados) de covid é de ', casos_por_milhao_cidade_selecionado, ' milhões de casos')
```

```
#Conferindo o número de mortes estaduais e municipais
mortes_estado = round(covid_estado_selecionados.query('is_last == True')['last_available_deaths'].sum()/1000,2)
mortes_cidade = round(covid_cidade.query('is_last == True')['last_available_deaths'].sum()/1000,2)
mortes_cidade_selecionado = round(covid_cidade_selecionados.query('is_last == True')['last_available_deaths'].sum()/1000,2)
print('O número de registros estaduais de mortes por covid é de', mortes_estado, 'mil mortes')
print('O número de registros municipais de mortes por covid é de', mortes_cidade, 'mil mortes')
print('O número de registros municipais (de municípios de estados selecionados) de mortes por covid é de', mortes_cidade_selecionado, 'mil mortes')
```

```
casos_e_mortes = pd.DataFrame({'Estadual':[casos_por_milhao_estado, mortes_estado],
                               'Municipal':[casos_por_milhao_cidade, mortes_cidade],
                               'Municipal de estados selecionados': [casos_por_milhao_cidade_selecionado, mortes_cidade_selecionado]})
casos_e_mortes = casos_e_mortes.rename(index={0:'casos [milhões]', 1:'mortes [mil]'})
casos_e_mortes
```

Como a diferença dos dados municipais, excluindo os estados em que a diferença de contagem é maior que 1%, acaba ficando muito grande optei por manter na base de dados de registros municipais todos os estados.

## 05 - Análise Nacional e Regional

```
#Plotagem do gráfico de novos casos de covid no Brasil
plt.figure(figsize=(20,16))
plt.subplot(2,1,1)
plt.grid(True)
plt.bar(covid_estado['date'].unique(), covid_estado.groupby('date').sum()['new_confirmed'], label='Novos Casos', color='y')
plt.plot(covid_estado.groupby('date').sum()['new_confirmed'].rolling(14).mean(), label='Média móvel de 14 dias', color='blue')
plt.plot(covid_estado['date'].unique(), covid_estado.groupby('date').sum()['new_deaths'], label='Mortes por covid', color='red')

plt.title('Novos casos de covid no Brasil', fontsize=20).set_color('b')
plt.legend().set_title('Legenda')
plt.xlabel('Data', fontsize=20)
plt.ylabel('Casos', fontsize=20)

#Plotagem do gráfico de novas mortes por covid no Brasil
plt.subplot(2,1,2)
plt.bar(covid_estado['date'].unique(), covid_estado.groupby('date').sum()['new_deaths'], label='Mortes por covid', color='r')
plt.plot(covid_estado.groupby('date').sum()['new_deaths'].rolling(14).mean(), label='Média móvel de 14 dias', color='blue')
plt.title('Mortes diárias por covid no Brasil', fontsize=20).set_color('b')
plt.legend().set_title('Legenda')
plt.xlabel('Data', fontsize=20)
plt.ylabel('Mortes', fontsize=20)
plt.show()
```

Iniciamos uma descida no número de casos a partir do mês de agosto e o gráfico nacional não mostra sinais de uma segunda onda, porém é muito cedo para afirmar algo e além disso se olharmos gráficos estaduais talvez uma segunda onda já esteja começando em alguns deles e está sendo mascarada no gráfico nacional por conta de estados que estejam com números descendentes. Nos gráficos europeus vemos claramente a segunda onda ocorrendo, porém para se fazer alguma comparação justa entre os países o ideal seria comparar regiões com densidade populacional semelhante, ao contrário do que muita gente faz que é comparar casos/milhão de habitantes, uma vez que o vírus se propaga de pessoa para pessoa e não de pessoa para pessoa em função do milhão de habitantes daquele país. O que quero dizer é que um país com alta concentração de pessoas por km² terá uma propagação muito mais veloz do que um país com a mesma população, espalhada numa área muito maior, já que as pessoas estarão menos aglomeradas. Desta forma, qualquer comparação tem suas restrições.

Coloquei, no gráfico de novos casos de covid, a linha com as mortes por covid apenas para deixar um pouco mais claro a ordem de grandeza do gráfico de mortes com relação ao de casos de covid.

Indica o número de pessoas que um paciente positivo para covid infecta. Caso a taxa esteja acima de 1 o número de novos casos aumenta, enquanto que abaixo de 1 significa que menos pessoas estão sendo infectadas do que o número de novos casos do dia anterior. A curva de casos é apenas o resultado da taxa de infecção. Pode-se observar que sempre que a taxa passa para baixo da linha horizontal vermelha (indicando o 1 do eixo Y esquerdo) a curva de novos casos de covid estará descendente e vice-versa.

Obs: foi utilizado os casos por semana epidemiológica somados de forma a retirar da curva o "erro" da falta de notificação que ocorre nos finais de semana.
